```
from sklearn.datasets import fetch_20newsgroups_vectorized

newsgroups_train = fetch_20newsgroups_vectorized(subset='train')
newsgroups_test = fetch_20newsgroups_vectorized(subset='test')
```

```
X_train = newsgroups_train.data
y_train = newsgroups_train.target

X_test = newsgroups_test.data
y_test = newsgroups_test.target
```

### Multinomial

```
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(X_train, y_train)

y_hat = clf.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_hat)
```

### Bernoulli

```
from sklearn.naive_bayes import BernoulliNB
clf = BernoulliNB()
clf.fit(X_train, y_train)

y_hat = clf.predict(X_test)

accuracy_score(y_test, y_hat)
```

### Complement

```
from sklearn.naive_bayes import ComplementNB
clf = ComplementNB()
clf.fit(X_train, y_train)

y_hat = clf.predict(X_test)

accuracy_score(y_test, y_hat)
```

### Gaussian

```
from sklearn.feature_extraction.text import TfidfTransformer
tfidf = TfidfTransformer().fit(X_train)
X_train = tfidf.transform(X_train)
X_test = tfidf.transform(X_test)
```

```
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train.toarray()[:100], y_train[:100])

y_hat = clf.predict(X_test.toarray())

accuracy_score(y_test, y_hat)
```

### Complement (tfidf)

```
from sklearn.naive_bayes import ComplementNB
clf = ComplementNB()
clf.fit(X_train, y_train)

y_hat = clf.predict(X_test)

accuracy_score(y_test, y_hat)
```
